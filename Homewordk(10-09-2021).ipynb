**25 .calculate the average word length of the following paragraph**

this is a paragraph which is written just for the purpose of providing content to let the average word length be calculated


```python
str="this is a paragraph which is written just for the purpose of providing content to let the average word length be calculated"
word=str.split()
length_of_word=0
Total_word=len(word)
for i in word:
  length_of_word=length_of_word+len(i)
print(f"The total number of words in the String is :{length_of_word}")
avg_Length=int(length_of_word/Total_word)
print(f"Average length of the word is :{avg_Length}")


```
